# Q2. Write a code to replace the odd numbers by -1 in the array created by using random.randint(1,50,20)

```python
arr1=np.random.randint(1,50,20)
```

```python
arr1
```

```python
arr1[arr1%2!=0]=-1
```

```python
arr1
```
